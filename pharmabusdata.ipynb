```python
#import dependencies

import pandas as pd
import matplotlib.pyplot as plt
import os
import json
import requests
```

```python
#read csv
df=pd.read_excel("./Business_dataset.xlsx")
df.head()
```

```python
#Data cleaning-dropna(), fillna(), replace
df=df.dropna()
df=df.rename(columns={""})
df.head()
```

```python
#Visualisations
#Bar plot of Brand Name on sum of Revenue
df_grouped=df.groupby("Brand Name").sum()["Revenue"]
yaxis=df_grouped.values
xaxis=df_grouped.index
df_grouped
plt.bar(xaxis, yaxis, color='r', alpha=0.5, align="center")
plt.title("Revenues by Drugs")
plt.xlabel("Drugs")
plt.ylabel("Revenue in Hundred Millions")
tick_locations = [value for value in xaxis]
plt.xticks(tick_locations,df_grouped.index, rotation=90)
plt.show()
```

```python
#Heatmap of purchases by Country
company_grouped=df.groupby("Company Name").sum()["Revenue"]
company_grouped=pd.DataFrame(company_grouped)
company_grouped=company_grouped.rename(index={"Centocor Ortho Biotech, Inc., Mitsubishi Tanabe Pharma": "Centocor Ortho Biotech", "Biogen Idec, Chugai Pharmaceutical, Genentech/Roche":"Chugai Pharmaceutical"})
```

```python
#Pie of company share in revenue terms
# yaxis=company_grouped["Revenue"]
# xaxis=company_grouped["Company Name"]
fig1=company_grouped["Revenue"].plot(kind="pie", title="Revenue by Pharma Companies", figsize=(25,8), autopct="%1.1f%%")
fig1.set_xlabel("Companies")
fig1.set_ylabel("Revenue in Hundred Millions")
```
